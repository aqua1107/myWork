```python
#R-2. 4

class Flower:

    def __init__(self, name, petals, price):
        self.name = name
        self.petals = petals
        self.price = price

    def get_name(self):
        return self.name

    def set_name(self, new_name):
        self.name = new_name

    def get_petals(self):
        return self.petals
    
    def set_petals(self, num):
        self.petals = num

    def get_price(self):
        return self.price
    
    def set_price(self, new_price):
        self.price = new_price

flower = Flower("Rose", 5, 2.99)

print(flower.get_name())
print(flower.get_petals())
print(flower.get_price()) 

flower.set_name("Tulip")
flower.set_petals(6)
flower.set_price(3.49)

print(flower.get_name())  
print(flower.get_petals())
print(flower.get_price())
```

```python
# R-2.5 ~ R-2.6

class CreditCard:
    
    def __init__(self, customer, bank, account, limit):
        self._customer = customer
        self._bank = bank
        self._account = account
        self._limit = limit
        self._balance = 0

    def get_customer(self):
        return self._customer

    def get_bank(self):
        return self._bank

    def get_account(self):
        return self._account

    def get_limit(self):
        return self._limit

    def get_balance(self):
        return self._balance

    def charge(self, price):
        if price + self._balance > self._limit:
            return False
        else:
            self._balance += price
            return True

    def make_payment(self, amount):
        try:
            if not isinstance(amount, (int, float)):
                raise ValueError("Payment amount must be a number.")

            elif amount < 0:
                raise ValueError("Payment amount cannot be negative.")
            
            elif amount > self._balance:
                self._balance = 0 
            else:
                self._balance -= amount

        except ValueError as e:
            print(f"Error: {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")


card = CreditCard("Alice", "BankX", "123456", 5000)

card.charge(100)

card.make_payment(50)

print(card._balance)

card.make_payment("fifty")

card.make_payment(-10)
```

```python
# R-2.7
class CreditCard:
    def __init__(self, customer, bank, account, limit, balance=0):
        self._customer = customer
        self._bank = bank
        self._account = account
        self._limit = limit
        self._balance = balance

    def get_customer(self):
        return self._customer

    def get_bank(self):
        return self._bank

    def get_account(self):
        return self._account

    def get_limit(self):
        return self._limit

    def get_balance(self):
        return self._balance

    def charge(self, price):
        if price + self._balance > self._limit:
            return False
        else:
            self._balance += price
            return True

    def make_payment(self, amount):
        self._balance -= amount
```
